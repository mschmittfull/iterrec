## Global params ##

```python
# path to data
basedir = '../data/'
#basedir = '/Users/mschmittfull/scratch_data/lss/ms_gadget/run4/00000400-01536-500.0-wig/'

sim_scale_factor = 0.625

# mesh options
Nmesh = 64

# cosmology of ms_gadget sims (to compute D_lin(z) so we can rescale true linear density)
# omega_m = 0.307494
# omega_bh2 = 0.022300
# omega_ch2 = 0.118800
# h = math.sqrt((omega_bh2 + omega_ch2) / omega_m) = 0.6774
cosmo_params = dict(Om_m=0.307494,
                   Om_L=1.0 - 0.307494,
                   Om_K=0.0,
                   Om_r=0.0,
                   h0=0.6774)
f_log_growth = np.sqrt(0.61826)

# Input catalogs
DMcat = Target(
    name='DM',
    in_fname=os.path.join(basedir, 'snap_%.4f_sub_sr0.0015_ssseed40400.bigfile' % sim_scale_factor),
    position_column='Position'
)

DMcat_small = Target(
    name='DM_small',
    in_fname=os.path.join(basedir, 'snap_%.4f_sub_sr0.00025_ssseed40400.bigfile' % sim_scale_factor),
    position_column='Position'
)

# Model meshs
z_rescalefac = linear_rescale_fac(current_scale_factor=1.0,
                                  desired_scale_factor=sim_scale_factor,
                                  cosmo_params=cosmo_params)
deltalin = Model(
    name='deltalin',
    in_fname=os.path.join(basedir, 'IC_LinearMesh_z0_Ng%d' % Nmesh),
    rescale_factor=z_rescalefac,
    read_mode='delta from 1+delta',
    filters=None,
    readout_window='cic')
```

## Do reconstruction ##

```python
# Reconstruction settings

smoothers = []
smoothers.append(GaussianSmoother(
    R1=20.0, # smoothing scale R in first iteration step
    R_reduction_factor=2.0, # factor by which R is reduced in each step.
    Rmin=1.0, # minimum smoothing scale allowed.
))

reconstructor = rec.Reconstructor(
    smoothers=smoothers,
    displacement_factor=1.0,
    Nsteps=4,
    Nmesh=Nmesh,
    paint_window='cic',
    name='A reconstruction test')
```

```python
# Load catalog and run reconstruction
#cat = DMcat.get_catalog()
cat = DMcat_small.get_catalog()
mesh_deltalin_rec = reconstructor.reconstruct_linear_density_from_catalog(cat)
```

## Measure reconstruction performance ##

```python
# Load true linear density
mesh_deltalin = deltalin.get_mesh()
```

```python
# Compute density of input catalg
mesh_delta_input, attrs = mass_weighted_paint_cat_to_delta(
    cat,
    weight=None,
    Nmesh=reconstructor.Nmesh,
    to_mesh_kwargs={
        'window': reconstructor.paint_window,
        'compensated': False,
        'interlaced': False},
    verbose=False)
```

```python
# Function to compute power spectrum
def calc_power(mesh, second=None, mode='1d', k_bin_width=1.0, verbose=False, los=None, BoxSize=None):
    """
    Basically the same as nbodykit FFTPower, but specify a certain binning in k.
    """
            
    if BoxSize is None:
        BoxSize = mesh.attrs['BoxSize']
    if len(BoxSize) > 1:
        assert BoxSize[0] == BoxSize[1]
        assert BoxSize[0] == BoxSize[2]
    boxsize = BoxSize[0]
    dk = 2.0 * np.pi / boxsize * k_bin_width
    kmin = 2.0 * np.pi / boxsize / 2.0

    if mode == '1d':
        res = FFTPower(first=mesh,
                        second=second,
                        mode=mode,
                        dk=dk,
                        kmin=kmin)
    elif mode == '2d':
        res = FFTPower(first=mesh,
                            second=second,
                            mode=mode,
                            dk=dk,
                            kmin=kmin,
                            poles=[0,2,4],
                            Nmu=5,
                            los=los)
    else:
        raise Exception("Mode not implemented: %s" % mode)

    return res
```

```python
# Compute power spectra of reconstructed and true linear density
BoxSize = cat.attrs['BoxSize']

# t: True linear density
# r: Reconstructed linear density
# i: Input catalog density
Ptt = calc_power(mesh_deltalin, BoxSize=BoxSize)
Prr = calc_power(mesh_deltalin_rec, BoxSize=BoxSize)
Prt = calc_power(mesh_deltalin_rec, second=mesh_deltalin, BoxSize=BoxSize)
Pii = calc_power(mesh_delta_input, BoxSize=BoxSize)
Pit = calc_power(mesh_delta_input, second=mesh_deltalin, BoxSize=BoxSize)
```

```python
# Plot power spectra
plt.figure(figsize=(6,4))
plt.loglog(Ptt.power['k'], Prr.power['power'], label='rr')
plt.loglog(Ptt.power['k'], Prt.power['power'], label='rt')
plt.loglog(Ptt.power['k'], Ptt.power['power'], label='tt')
plt.loglog(Ptt.power['k'], Pii.power['power'], label='ii')

plt.legend(loc='best', fontsize=14);
plt.xlabel(r'$k$ [h/Mpc]');
plt.ylabel(r'$P$ [$($Mpc/h$)^3$]');
```

```python
# Plot correlation coefficient
plt.figure(figsize=(6,4))
plt.semilogx(Ptt.power['k'], Prt.power['power']/np.sqrt(Ptt.power['power']*Prr.power['power']),
            label='X=reconstructed')

plt.semilogx(Ptt.power['k'], Pit.power['power']/np.sqrt(Pii.power['power']*Ptt.power['power']),
            label='X=input')

plt.legend(loc='best', fontsize=14);
plt.xlabel(r'$k$ [h/Mpc]');
plt.ylabel(r'$r_{cc}(\delta_\mathregular{lin},X)$');
```

```python
# Plot 1 - correlation coefficient^2
plt.figure(figsize=(6,4))
plt.loglog(Ptt.power['k'], 1.0 - Prt.power['power']**2/(Ptt.power['power']*Prr.power['power']),
            label='X=reconstructed')
plt.loglog(Ptt.power['k'], 1.0 - Pit.power['power']**2/(Ptt.power['power']*Pii.power['power']),
            label='X=input')

plt.legend(loc='best', fontsize=14);
plt.xlabel(r'$k$ [h/Mpc]');
plt.ylabel(r'$1-r^2_{cc}(\delta_\mathregular{lin},X)$');
```
